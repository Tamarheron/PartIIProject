```python
def plotSingle(data, title):
    weights1=[1]
    AB = np.array([data[5],
                data[10],data[15],
                data[20],data[25]])

    xs=AB[:,0]
    ys=AB[:,1:]

    humanAB=np.average(ys[:,[0]],axis=1,weights=weights1)
    fixedAB=np.average(ys[:,[1]],axis=1,weights=weights1)
    variableAB=np.average(ys[:,[2]],axis=1,weights=weights1)

    optimalAB=[45.5,95.5,145.5,195.5,245.5]
    humanABregret=(optimalAB-humanAB)/optimalAB
    fixedABregret=(optimalAB-fixedAB)/optimalAB
    variableABregret=(optimalAB-variableAB)/optimalAB

    width = 0.5
    fig, [ax1,ax2] = plt.subplots(nrows=1,ncols=2, figsize=(10,5))
    human=ax1.bar(xs-width*1.5,humanABregret,width=width,color='grey')
    fixed=ax1.bar(xs-width*.5,fixedABregret,width=width, color='lightsteelblue')
    variable=ax1.bar(xs+width*.5,variableABregret,width=width,color='blue')

    plt.ylabel("Normalized Regret")
    plt.xlabel("Trajectory Length")

    red_patch = mpatches.Patch(color='blue', label='Variable Beta Robot')
    blue_patch = mpatches.Patch(color='lightsteelblue', label='Fixed Beta Robot')
    grey_patch = mpatches.Patch(color='grey', label='Human')

    plt.title("regret "+title)
    plt.legend(handles=[grey_patch,blue_patch,red_patch],loc=3)
    plt.savefig("regret "+title)
    

    human=ax2.bar(xs-width*1.5,humanAB,width=width,color='grey')
    fixed=ax2.bar(xs-width*.5,fixedAB,width=width, color='lightsteelblue')
    variable=ax2.bar(xs+width*.5,variableAB,width=width,color='blue')
    plt.ylabel("Score")
    plt.xlabel("Trajectory Length")

    red_patch = mpatches.Patch(color='blue', label='Variable Beta Robot')
    blue_patch = mpatches.Patch(color='lightsteelblue', label='Fixed Beta Robot')
    grey_patch = mpatches.Patch(color='grey', label='Human')

    plt.title("Score "+title)
    plt.legend(handles=[grey_patch,blue_patch,red_patch],loc=2)
    plt.savefig("Score "+title)
```
